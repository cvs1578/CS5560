```
#we need to load the dataset. We will use the read.csv module. 
#The inferSchema parameter provided will enable Spark to automatically determine the data type for each column but it has to go over the data once.
# If you donâ€™t want that to happen, then you can instead provide the schema explicitly in the schema parameter.

df = spark.read.csv("/content/drive/MyDrive/Colab Notebooks/ICP4/data.csv", header=True, inferSchema= True) #inferSchema keeps data type
df.printSchema()
```

TASK: Perform 3 transformations on data.csv file

```
df.show()
```

```
#With the filter command we can show specific results 
df.filter("Partner == 'Yes'").show() #This function will only show the dataset if they are a partner.
```

```
#groupBy will collect data from identical groups and perform functions on them. Note that this can be considered both a transformation and action
#First you define the group you want to 'compress'/analyze (for instance gender)
#Then you decide which numerical function to use (for this I used mean) then define what group you want to caluclate (so monthly charges)
df.groupBy("gender").mean("MonthlyCharges").show() #This groups all the different genders together and caluclates the gender's average monthly charge
```

```
#dropDuplicates removes any duplicates in the dataset. This is very useful in making sure you don't accidentally show the same data twice.
df.select("customerID").dropDuplicates().show()
```

Task: Perform 3 actions on the file 

```
#Count counts the amount of elements/rows in the dataset. This is very useful in identifying the impact your transforms have on the data
print(df.count())
print(df.filter("Partner == 'Yes'").count())
```

```
#take will return the first k elements. This is so you can choose data points and mess around with them without ruining the data.
dft=df.select("MonthlyCharges").take(5)
for x in dft:
  print(x)
```

```
#collect returns the dataset as a list of rows. It is great for allowing computer scientists the ability to fit the entire dataset into the program's memory
df.collect()
```
