Now we're gonna tokenize the text

```

paragraph = "Xi Jinping is a Chinese politician who has served as General Secretary of the Chinese Communist Party (CCP) and Chairman of the Central Military Commission (CMC) since 2012, and President of the People's Republic of China (PRC) since 2013. He has been the paramount leader of China, the most prominent political leader in the country, since 2012. The son of Chinese Communist veteran Xi Zhongxun, he was exiled to rural Yanchuan County as a teenager following his father's purge during the Cultural Revolutionandlived in a cave in the village of Liangjiahe, where he joined the CCP and worked as the party secretary."
tokenwords = nltk.word_tokenize(paragraph) #This tokenizes the paragraph, turning the string into a list of each word.
tokenwords
```

Part of Speech (POS) tagger

```
POS = nltk.pos_tag(tokenwords) #This adds another column to the list to identify each word
POS
```

Name Entity Recognizer (NER)

```
NER = nltk.chunk.ne_chunk(POS) #This adds identifiers for the nouns
print(NER)
```

Lemmatization

```
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for x in tokenwords:
  print ("{} is {}".format(x, lemmatizer.lemmatize(x))) #Lemmatize each word individually for user to see
```

Parsing

```
grammar = "NP:{<DT>?<JJ>*<NN>}" #Grammar definition. To be more precise = more defined. This is just basic for regexparser
rd = nltk.RegexpParser(grammar)
output = rd.parse(POS)
print(output)
#output.draw() ##NOTE: Since draw doesn't work on Colab we are unable to view the tree
```
