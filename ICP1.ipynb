Readind the files as a pandas dataframe 

```
dataset=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ICP1/input_file.txt",delimiter="\t")
dataset
```

converting the data frame to text

```
text= dataset.to_string()
text
```

Breaking the text into sentences 

```
tokenized_text = sent_tokenize(text)
print(tokenized_text)
```

Breaking the text into words 

```
tokenized_word=word_tokenize(text)
print(tokenized_word)
print(len(tokenized_word))
```

finding the frequencies distribution

```
fdist = FreqDist(tokenized_word)
print(fdist)
```

Finding the two most common words  

```
fdist.most_common(2)
```

Plot of the frequencies distribution

```
v=fdist.plot(75,cumulative=False)
plt.show()
```

Breaking the text into letters

```
char_text=[]
char_text[:0]= text #This converts the entire string into the individual characters in a list
print(char_text)
```

finding the frequency distribution of the characters

```
cfdist = FreqDist(char_text)
print(cfdist)
```

Finding the two common letters

```
cfdist.most_common(2)
```

Ploting frequency distribution

```
v1=cfdist.plot(403,cumulative=False)
plt.show()
```

Finding only the vowels

```
char_vow=[]
for x in char_text:
  if x == 'a' or x == 'e' or x == 'i' or x =='o' or x =='u' or x =='A' or x =='E' or x =='I' or x =='O' or x =='U':
    char_vow[:0] = x
print(char_vow) #Note: This data is backwards
```

finding the frequency distribution of the vowels

```
cvdist = FreqDist(char_vow)
print(cvdist)
```

Finding the two common vowels

```
cvdist.most_common(2)
```

Ploting frequency distribution

```
v2=cvdist.plot(128,cumulative=False)
plt.show()
```

Let's remove the noise from the data

```
NoNoise = [] #This will hold all the words that aren't noise
Noise = True #According to the data, the real words are within [ ], so we'll be using that to determine if noise is present
for x in tokenized_word:
  if x == "]": 
    Noise = True
  if Noise == False: #
    NoNoise.append(x) #append leaves the format in string.
  if x == "[":
    Noise = False
print(NoNoise)
```
