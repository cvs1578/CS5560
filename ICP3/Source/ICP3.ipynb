```
Text1 = "She took the midnight train going anywhere"
Text2 = "I can show you the world"
Text3 = "Did you ever hear the tragedy of Darth Plagueis The Wise?"
```

```
#We are going to use Spacy in combination with Textacy since textacy has a verb_object_triples function
import spacy
import textacy

#First, we must tokenize the wprd
nlp = spacy.load('en')
sent1 = nlp(Text1)
for tok in sent1: 
  print(tok.text, "-->",tok.dep_,"-->", tok.pos_)
sent2 = nlp(Text2)
for tok in sent2: 
  print(tok.text, "-->",tok.dep_,"-->", tok.pos_)
sent3 = nlp(Text3)
for tok in sent3: 
  print(tok.text, "-->",tok.dep_,"-->", tok.pos_)
```

```
#Next, we will just use the function in textacy that will output a 
Trip1 = textacy.extract.subject_verb_object_triples(sent1)
for x in Trip1:
  print (x)
Trip2 = textacy.extract.subject_verb_object_triples(sent2)
for x in Trip2:
  print (x)
Trip3 = textacy.extract.subject_verb_object_triples(sent3)
for x in Trip3:
  print (x)
```

```
import nltk
nltk.download('all')
from nltk.corpus import wordnet
```

```
word = wordnet.synsets("dog")
print("Hyponym:",word[0].hyponyms())
print("Hypernym:",word[0].hypernyms())
print("Meronym:",word[0].part_meronyms()) #Has part, substance, or member.
print("Holonym:",word[0].member_holonyms()) #Has part, substance, or member. 
print("Entailment:",word[0].entailments())
```
